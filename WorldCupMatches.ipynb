# DATA CLEANING --------------------------------------------->>>

```python
df.isnull() # Shows the null values and returns Boolean values
# Where 'True' means NULL value is present
```

```python
# Shows the sum of null values in each column
missing_value = df.isnull().sum()
missing_value
```
